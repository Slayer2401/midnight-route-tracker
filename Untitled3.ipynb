```python
# üö¶ Efficient Traffic Object Detection using YOLOv8-Tiny
# Google Colab Notebook

# ===============================
# Step 1: Install YOLO
# ===============================
!pip install ultralytics

# ===============================
# Step 2: Import YOLO library
# ===============================
from ultralytics import YOLO

# ===============================
# Step 3: Upload Dataset (Roboflow)
# ===============================
# Go to Roboflow ‚Üí download dataset in "YOLOv5 PyTorch" format
# In Colab, click the folder icon (left sidebar) ‚Üí upload your ZIP
# Example file name: traffic_dataset.zip
import zipfile

zip_path = "/content/traffic-flow-counting.v2i.yolov5pytorch.zip"  # Change if different
with zipfile.ZipFile(zip_path, 'r') as zip_ref:
    zip_ref.extractall("/content/traffic_dataset")

# ===============================
# Step 4: Train YOLOv8-Tiny
# ===============================
model = YOLO("yolov8n.pt")   # nano/tiny model (fast, small size)
model.train(
    data="/content/traffic-flow-counting.v2i.yolov5pytorch.zip",
    epochs=10,         # quick training
    imgsz=320          # lower image size = faster
)

# ===============================
# Step 5: Test Model (Prediction)
# ===============================
# Run detection on test images
results = model.predict(
    source="/content/traffic-flow-counting.v2i.yolov5pytorch.zip",
    save=True
)

# ===============================
# Step 6: Show one result
# ===============================
import matplotlib.pyplot as plt
import cv2

# Replace filename with one from your dataset
img_path = results[0].save_dir + "/image1.jpg"

img = cv2.imread(img_path)
plt.imshow(cv2.cvtColor(img, cv2.COLOR_BGR2RGB))
plt.axis("off")
plt.show()

# ===============================
# Done üéâ
# ===============================
# Outputs are saved in /content/runs/detect/predict/
# You can download them by right-clicking the folder in sidebar ‚Üí Download
```
